```python
import pandas as pd
import numpy as np
import csv
import sys  
sys.path.append('/home/ndsviriden/Minimax94/src/utils')  
from Downloader import ExtractorData
from os import listdir
from os.path import isfile, join, isdir
from tqdm import tqdm_notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
extractor = ExtractorData()
step = 25
path = '/mnt/HARD/Minimax94/data/data_all/CSV/'
path_raw = path + 'Raw_extended/'
stations_all = pd.read_csv(path + 'station_def.csv', 
                           usecols=['station_id'])
stations_all = list(stations_all['station_id'].unique())

stations_in_dir = [int(f.split('_', 1)[0]) for f in listdir(path_raw) 
                   if isfile(join(path_raw, f)) if int(f.split('_', 1)[0])]
stations_in_dir.sort()
print('Stations in directory: ', stations_in_dir)

stations_in_dir = set(stations_in_dir)
stations_to_extract = [st for st in stations_all if st not in stations_in_dir]
sensor_list=[1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 12, 13, 14, 16, 17, 31]
```

```python
sensor_list = [1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 12, 13, 14, 16, 17, 31]
for i in tqdm_notebook(range(0, len(stations_to_extract), step)):
    stations = stations_to_extract[i: i+step]
    print('Extracting stations: ', stations)
    df_raw = extractor.ExtractFromDB(stations, '2007-01-01', '2017-01-01', sensor_list)
    
    for station in stations:
        print('Writing data into ' + path_raw + str(station) + '_raw.csv')
        df_raw[df_raw['station_id']==station].to_csv(path_raw + str(station) + '_raw.csv', index=False)
```
