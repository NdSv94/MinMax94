## Adding timezone information to station_def

```python
db = MySQLdb.connect("192.168.1.230", 
                     user='lm_ro', 
                     passwd='lm_rovtntj',
                     db='cup_system3',
                     port=3306)

# utf-8 encoding
db.set_character_set("utf8")
dbc = db.cursor()

# SELECT from cyrillic fields
dbc.execute("SELECT * FROM objects;")

# getting the results
data = dbc.fetchall()

# getting the column names
dbc = db.cursor()
dbc.execute("SELECT COLUMN_NAME FROM INFORMATION_SCHEMA.COLUMNS WHERE TABLE_NAME = 'objects';")
columns = dbc.fetchall()
columns = tuple([column[0] for column in columns])

# creating pandas table with column names from above
df_info = pd.DataFrame(list(data), columns=columns)

#df.to_csv('/home/ndsviriden/data_csv/station_def.csv')
```

## Get timezone for station

```python
def parse_timezone(s):
    p_re = '(?<="raw_timezone":"\+)[0-9]{1,2}'
    p = re.findall(p_re, s)
    if p:
        timezone = int(p[0])
        if timezone >= 12:
            print(timezone)
        else:
        timezone = 3
    return timezone
```

```python
df_info['timezone'] = df_info['properties'].apply(parse_timezone)
```

## Add timezone column into station_def table

```python
df_timezone = pd.merge(df, df_info[['id', 'timezone']], how='left', on='id', left_on=None, right_on=None,
         left_index=False, right_index=False, sort=True, copy=True, indicator=False)
df_timezone.loc[df_timezone['timezone'].isnull(), 'timezone'] = 3
df_timezone
```

```python
df_timezone = df_timezone.rename(columns={'id': 'station_id'})
```

## Writing the result table into file
