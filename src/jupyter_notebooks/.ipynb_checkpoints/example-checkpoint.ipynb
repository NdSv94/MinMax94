```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
import csv
from copy import deepcopy
import random
import gc
from Preprocessing import Preprocessor
from os import listdir
from os.path import isfile, join, isdir
from tqdm import tqdm_notebook
import sys
sys.path.append('/home/ndsviriden/MinMax94/src/utils')
import Preprocessing
import warnings
warnings.filterwarnings('ignore')
```

```python
# getting file tree in directory "data_csv", which contains raw unfiltered data
mypath = '/home/ndsviriden/data_csv/Raw'
dateparse = pd.to_datetime
station_files = [f for f in listdir(mypath) if isfile(join(mypath, f)) if int(f.split('_', 1)[0])]
station_files.sort()

# reading loaded csv files from data_csv directory, output is a list (length=number of stations) of raw df 
raw_lmeteo_list = [pd.read_csv(mypath + '/' + station, index_col=0, 
                    dtype={'station_id': int,'date_time': str}, 
                    date_parser = dateparse, parse_dates = ['date_time']) 
                    for station in tqdm_notebook(station_files[:4])]

#fix problems with indexing, later this will be solved in loading part
for elem in raw_lmeteo_list:
    elem.reset_index(drop=True, inplace=True)

raw_data = pd.concat(raw_lmeteo_list)

# adding generated ids and valid column
# generated ids will be removed by real ids from db, this is done only for testing
raw_data['id'] = random.sample(range(len(raw_data)), len(raw_data))
raw_data['valid'] = True
raw_data = raw_data.reset_index(drop=True)

del raw_lmeteo_list
gc.collect()

raw_data.head()
```

```python
# preprocessor module
preprocessor = Preprocessor()

# put sensor data from different sensor types (t_air, t_road, pressure, etc.) into different columns
meteo_split = preprocessor.PivotTable(raw_data)

# sometimes pressure data is saved as mm_Hg and sometimes as 10*mm_Hg (760 and 7600)
# fixing this mismaching
meteo_split = preprocessor.FixPressureScale(meteo_split) 

meteo_split.head(10)
```

```python
# split table into patterns
# if the gap between two records > '2h' then partitioning takes place
# if the length of the pattern < '12h' then pattern is removed
patterns_list = preprocessor.CreatePatternList(meteo_split)

# interpolating data from patterns into 30min timestamps
patterns_interpolated = preprocessor.InterpolatePatterns(patterns_list)
patterns_interpolated = [pattern for pattern in patterns_interpolated if len(pattern > 0)]
result_data = pd.concat(patterns_interpolated)
```

```python
def plot_data(df=result_data, station=113, start=pd.Timestamp(2012, 1, 1), 
              end=pd.Timestamp(2017, 1, 1), sensors=['t_road']):
    
    df_plot = df[df['station_id']==station]
    df_plot = df_plot[(df_plot['date_time'] <= end) & (df_plot['date_time'] >= start)]
    
    color = ['b', 'g', 'm']
    i = 0
    plt.figure(figsize=(30, 10))
    for sensor in sensors:
        plt.plot_date(df_plot.date_time, df_plot[('data', sensor)], color[i],
                     linestyle='none', marker='o', label=sensor, markersize=6)

        i+=1
        plt.title(sensor, fontsize=24)
        plt.legend(loc='upper right', fontsize=14)
    plt.grid(which='both')
    plt.show()
    pass
```
